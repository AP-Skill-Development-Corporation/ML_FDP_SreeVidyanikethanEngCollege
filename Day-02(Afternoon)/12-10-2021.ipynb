### Unsupervised Learning
- You don't need to supervise the model. Instead we need to allow the machine to work on its own.
    - Clustering: Used to identify the similar groups in dataset.
    
#### K-Means Clustering Model
- Easy to classify the dataset.
- K Means takes a dataset with constant values K and returns K no of clusters.

#### How does the K Means will work.
- Step1: Select k no of clusters randomly( K = 2 data points)
- Step2: Calculate distance
- Step3: Calculate new clusters
- Step4: Divide into groups
- Step5: Repeat the same procedure

```python
#sno   X   Y
#-----------------
# 1   1     1 
# 2   1.5   2
# 3   3     4
# 4   5     7
# 5   3.5   5
# 6   4.5   5
# 7   3.5   4.5
```

- Step1: Select K
    - Let K = 2
        m1 = (1,1) and m2 = (5,7)
        
- Step2: Calculate distance
    - Sqrt((X2 - X1)^2 + (Y2 - Y1)^2)

```python
import math
```

```python
## Distance between (1,1), (1,1) and (5,7) ,(1,1)
##                (x1,y1) (x2,y2) & (x1,y1),(x2,y2)
```

```python
print(math.sqrt((1-1)**2 + (1-1)**2))
print(math.sqrt((5-1)**2 + (7-1)**2))
```

```python
## Distance between (1.5,2), (1,1) and (1.5,2) ,(5,7)
##                (x1,y1) (x2,y2) & (x1,y1),(x2,y2)
```

```python
print(math.sqrt((1-1.5)**2 + (1-2)**2))
print(math.sqrt((5-1.5)**2 + (7-2)**2))
```

```python
# Updated values of X and Y

#        Distance     Distance
#sno     of m1         of m2
#-----------------------------
#1        0       <     7.211
#2       1.11     <     6.10
#3       3.60     <=     3.60
#4       7.21     >       0
#5       4.71     >     2.5
#6       5.31     >     2.06
#7       4.30     >     2.91
```

- Step3: Calculate new clusters

- Cluster1
    - K1(X) = 1,1.5,3 and K1(Y) = 1,2,4
    - Find mean
        - k1(X) = (1 + 1.5 + 3)/3 = 1.8333
        - K1(Y) = (1 + 2 + 4)/3  = 2.333
- Cluster2
    - K2(X) = 5,3.5,4.6,3.5 and K2(Y) = 7,5,5,4.5
    - Find mean
         - k2(X) = (5+3.5+4.5+3.6)/4 = 4.15
         - K2(Y) = (7 + 5+5+4.5)/4  = 5.375

```python
(1 + 1.5 + 3)/3
(1 + 2 + 4)/3
(5+3.5+4.5+3.6)/4
(7 + 5+5+4.5)/4
```

**Updated K values are**
- m1 = (1.833,4.15) and m2 = (2.33,5.375)

Step5: Repeat the same procedure
- Until either mean values of iterations will be same or No changes in cluster of X and Y

### Dimentionality Reduction
- Dimensions are represented as columns or variables or features in dataset.
- Dimentionality Reduction is commonly used to reduce the no of dimensions(independent variable or features)
- Two types of Dimentionality Reduction
    1. Feature elimination: Removing unnecessary features 
    2. Feature extraction: We create a new column where ecah new column is a combination of old columns
        - Eg: If we have 2 columns --> Car_Cost, Car_EMI
            - Car_Price = Car_Cost + Car_EMI

**Feature extraction**
1. Principle Component Analysis
2. Independent Component analysis
3. Random projections
4. etc,...

```python
df1.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_transform = scaler.fit_transform(X)
x_transform
```

```python
pca = PCA()
pca
```

```python
pca = PCA(n_components = 2)
pca
```

```python
new_col = pca.fit_transform(x_transform)
new_col
```
